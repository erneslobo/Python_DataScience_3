```python
import numpy as np
```

```python
x = np.random.randint(low=1, high=100, size=100)
y = np.random.randint(low=1, high=100, size=100)
Exy = (x*y).sum()
N = x.size
Ex = x.sum()
Ey = y.sum()

Ex2 = (x**2).sum()
Ex_2 = (x.sum())**2
den = N*Ex2 - Ex_2
numA = N*Exy - Ex*Ey
numB = Ex2*Ey - Ex*Exy
a = numA / den
b = numB / den
print(a)
print(b)


```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.scatter(x,y)
plt.title("Title")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
